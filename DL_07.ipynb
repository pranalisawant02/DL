Interpreting Tree-Based Models

```python
# Step 1: Import libraries
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier, plot_tree
import pandas as pd

# Step 2: Load dataset
iris = load_iris()
X = iris.data
y = iris.target

# Convert to DataFrame (optional, for better understanding)
df = pd.DataFrame(X, columns=iris.feature_names)
df['target'] = y

# Step 3: Train Decision Tree
model = DecisionTreeClassifier(criterion='gini', max_depth=3)
model.fit(X, y)

# Step 4: Visualize Decision Tree
plt.figure(figsize=(12,8))
plot_tree(model,
          feature_names=iris.feature_names,
          class_names=iris.target_names,
          filled=True)
plt.title("Decision Tree Visualization")
plt.show()

# Step 5: Feature Importance
importance = model.feature_importances_

# Display feature importance
for i, v in enumerate(importance):
    print(f"{iris.feature_names[i]}: {v:.4f}")

# Step 6: Plot Feature Importance
plt.barh(iris.feature_names, importance)
plt.xlabel("Importance Score")
plt.title("Feature Importance")
plt.show()
```

Random Forest Interpretaion

```python
# Step 1: Import libraries
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.ensemble import RandomForestClassifier
import pandas as pd
import numpy as np

# Step 2: Load dataset
iris = load_iris()
X = iris.data
y = iris.target

df = pd.DataFrame(X, columns=iris.feature_names)
df['target'] = y

# Step 3: Train Random Forest
model = RandomForestClassifier(n_estimators=100, random_state=42)
model.fit(X, y)

# Step 4: Feature Importance (GLOBAL interpretation)
importance = model.feature_importances_

# Sort features by importance
indices = np.argsort(importance)

# Display feature importance
print("Feature Importance:")
for i in indices:
    print(f"{iris.feature_names[i]}: {importance[i]:.4f}")

# Step 5: Plot Feature Importance
plt.figure(figsize=(8,5))
plt.barh(range(len(indices)), importance[indices])
plt.yticks(range(len(indices)), [iris.feature_names[i] for i in indices])
plt.xlabel("Importance Score")
plt.title("Random Forest Feature Importance")
plt.show()
```
